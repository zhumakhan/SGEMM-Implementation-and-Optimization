```python
sizes = ['conv1','conv2','conv3','conv4','conv5','conv6']
kernel_names = ['prefetching','cublass']
K=1000

table_old = [
    [
        340.46, 616, 621, 632, 616, 495 
    ],
    [
        1.0863*K,523, 395.3, 331.40, 277.26, 374.76
    ],
]

table = [
    [
        338, 515, 522, 545, 580, 456 
    ],
    [
        1.0773*K,540, 391.3, 330.40, 276.26, 375.76
    ],
]


import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {'kernel':[],'size':[],'time (us)':[]}

for i,kernel in enumerate(kernel_names):
  for j,size in enumerate(sizes):
    data['kernel'].append(kernel)
    data['size'].append(size)
    if  table[i][j]!=0:
      data['time (us)'].append(table[i][j])
    else:
      data['time (us)'].append(None)
data
res = pd.DataFrame.from_dict(data)

fig,ax = plt.subplots(figsize=(15,10))
sns.pointplot(data=res[['kernel','size','time (us)']],
             y='time (us)',x='size',
             hue='kernel',ax=ax)
ax.set(title='size vs time (us) for each kernel')
plt.show()

```
